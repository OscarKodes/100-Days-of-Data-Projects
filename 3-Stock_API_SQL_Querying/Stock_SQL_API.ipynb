# Imports for api keys

```python
from dotenv import load_dotenv
import os

load_dotenv()
```

# Get Data

- Send a SQL Query request to API

```python
import http.client

conn = http.client.HTTPSConnection("hotstoks-sql-finance.p.rapidapi.com")

payload = "\"SELECT * FROM stocks WHERE symbol in ('GOOG', 'GME', 'LXRX') ORDER BY price_change_percent_1m DESC\""
# payload = "\"SELECT symbol, name FROM stocks ORDER BY price_change_percent_1m DESC\""


headers = {
    'content-type': "text/plain",
    'X-RapidAPI-Key': os.getenv("api_key"),
    'X-RapidAPI-Host': "hotstoks-sql-finance.p.rapidapi.com"
    }

conn.request("POST", "/query", payload, headers)

res = conn.getresponse()
data = res.read()

print(data.decode("utf-8"))
```

# Convert Received JSON String to Dictionary

```python
import json

s = "" + data.decode("utf-8")

d = json.loads(s)

results = d['results']

print(len(results))

print(type(results))
```

# Convert Dictionary to Pandas DataFrame

```python
import pandas as pd

df = pd.DataFrame(results)

# make sure all columns and rows are shown
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

df.head()
```
