# Practice Window Functions

"A window function performs a calculation across a set of table rows that are somehow related to the current row. This is comparable to the type of calculation that can be done with an aggregate function. But unlike regular aggregate functions, use of a window function does not cause rows to become grouped into a single output row â€” the rows retain their separate identities. Behind the scenes, the window function is able to access more than just the current row of the query result."

https://mode.com/sql-tutorial/sql-window-functions/

### SUM

- We get the SUM total over all the amounts paid.
- We can choose to PARTITION BY a category like "customer_id"
- ORDER BY is specified so SQL knows in which order to create the sum totals.
    - In this case we use "payment_date"

```python
%%sql
    
SELECT c.customer_id,
        first_name,
        last_name,
        payment_id,
        payment_date,
        amount,
        SUM(amount) OVER 
        (PARTITION BY c.customer_id ORDER BY payment_date) 
        AS running_total
FROM customer c
JOIN payment p
ON c.customer_id = p.customer_id
WHERE c.customer_id < 3;
```

### SUM, AVG, COUNT

- We get the SUM total over all the amounts paid.
- We can choose to PARTITION BY a category like "customer_id"
- ORDER BY is specified so SQL knows in which order to create the sum totals.
    - In this case we use "payment_date"

```python
%%sql
    
SELECT c.customer_id,
        first_name,
        last_name,
        payment_id,
        payment_date,
        amount,
        COUNT(payment_id) OVER
            (PARTITION BY c.customer_id ORDER BY payment_date) 
            AS running_count,
        SUM(amount) OVER 
            (PARTITION BY c.customer_id ORDER BY payment_date) 
            AS running_sum,
        AVG(amount) OVER
            (PARTITION BY c.customer_id ORDER BY payment_date) 
            AS running_avg
FROM customer c
JOIN payment p
ON c.customer_id = p.customer_id
WHERE c.customer_id < 3;
```

### ROW NUMBER

"ROW_NUMBER() does just what it sounds likeâ€”displays the number of a given row. It starts are 1 and numbers the rows according to the ORDER BY part of the window statement. ROW_NUMBER() does not require you to specify a variable within the parentheses."

```python
%%sql

SELECT film_id,
        title,
        rating,
        ROW_NUMBER() OVER 
            (ORDER BY film_id)
            AS row_number
FROM film
WHERE title LIKE '%a';
    
```

"Using the PARTITION BY clause will allow you to begin counting 1 again in each partition. The following query starts the count over again for each terminal."

```python
%%sql

SELECT film_id,
        title,
        rating,
        ROW_NUMBER() OVER 
            (PARTITION BY rating
             ORDER BY film_id)
            AS row_number
FROM film
WHERE title LIKE '%a';
```

### RANK

"RANK() is slightly different from ROW_NUMBER(). If you order by start_time, for example, it might be the case that some terminals have rides with two identical start times. In this case, they are given the same rank, whereas ROW_NUMBER() gives them different numbers. In the following query, you notice the 4th and 5th observations for start_terminal 31000â€”they are both given a rank of 4, and the following result receives a rank of 6."

```python
%%sql

SELECT film_id,
        title,
        rating,
        length,
        RANK() OVER 
            (PARTITION BY rating
             ORDER BY length)
            AS rank
FROM film
WHERE title LIKE '%y';
```

### DENSE RANK

You can also use DENSE_RANK() instead of RANK() depending on your application. Imagine a situation in which three entries have the same value. Using either command, they will all get the same rank. For the sake of this example, let's say it's "2." Here's how the two commands would evaluate the next results differently.

RANK() would give the identical rows a rank of 2, then skip ranks 3 and 4, so the next result would be 5
DENSE_RANK() would still give all the identical rows a rank of 2, but the following row would be 3â€”no ranks would be skipped.

```python
%%sql

SELECT film_id,
        title,
        rating,
        length,
        DENSE_RANK() OVER 
            (PARTITION BY rating
             ORDER BY length)
            AS dense_rank
FROM film
WHERE title LIKE '%y';
```

### NTILE

You can use window functions to identify what percentile (or quartile, or any other subdivision) a given row falls into. The syntax is NTILE(*# of buckets*). In this case, ORDER BY determines which column to use to determine the quartiles (or whatever number of 'tiles you specify).

```python
%%sql

SELECT film_id,
        title,
        rating,
        length,
        NTILE(4) OVER
         (PARTITION BY rating 
          ORDER BY length)
          AS quartile,
        NTILE(5) OVER
         (PARTITION BY rating 
          ORDER BY length)
          AS quintile,
        NTILE(100) OVER
         (PARTITION BY rating 
          ORDER BY length)
          AS percentile
FROM film
WHERE title LIKE '%y';
```

```python
%%sql

SELECT film_id,
        title,
        rating,
        length,
        NTILE(4) OVER
         (ORDER BY length)
          AS quartile,
        NTILE(5) OVER
         (ORDER BY length)
          AS quintile,
        NTILE(100) OVER
         (ORDER BY length)
          AS percentile
FROM film
WHERE rating = 'G';
```

### LAG and LEAD

"It can often be useful to compare rows to preceding or following rows, especially if you've got the data in an order that makes sense. You can use LAG or LEAD to create columns that pull values from other rowsâ€”all you need to do is enter which column to pull from and how many rows away you'd like to do the pull. LAG pulls from previous rows and LEAD pulls from following rows."

```python
%%sql

SELECT film_id,
        title,
        rating,
        length,
        LAG(length, 1) OVER
            (PARTITION BY rating ORDER BY length) 
            AS "lag_(previous_row_length)"
FROM film
WHERE title LIKE '%y';
```

```python
%%sql

SELECT film_id,
        title,
        rating,
        length,
        LEAD(length, 1) OVER
            (PARTITION BY rating ORDER BY length) 
            AS "Lead_(next_row_length)"
FROM film
WHERE title LIKE '%y';
```

**"Lag and Lead are especially useful if you want to calculate differences between rows."**

```python
%%sql

SELECT film_id,
        title,
        rating,
        length,
        length -LAG(length, 1) OVER
            (PARTITION BY rating ORDER BY length)
            AS difference
FROM film
WHERE title LIKE '%y';
```

**"The first row of the difference column is null because there is no previous row from which to pull. Similarly, using LEAD will create nulls at the end of the dataset. If you'd like to make the results a bit cleaner, you can wrap it in an outer query to remove nulls."**

```python
%%sql

SELECT *
FROM (
    SELECT film_id,
            title,
            rating,
            length,
            length -LAG(length, 1) OVER
                (PARTITION BY rating ORDER BY length)
                AS difference
    FROM film
    WHERE title LIKE '%y'
    ) sub
WHERE sub.difference IS NOT NULL;
```

### Defining a window alias

"If you're planning to write several window functions in to the same query, using the same window, you can create an alias. Take the NTILE example above."

**The WINDOW clause, if included, should always come after the WHERE clause.**

```python
%%sql

SELECT film_id,
        title,
        rating,
        length,
        NTILE(4) OVER ntile_window AS quartile,
        NTILE(5) OVER ntile_window AS quintile,
        NTILE(100) OVER ntile_window AS percentile
FROM film
WHERE title LIKE '%y'
WINDOW ntile_window
    AS (PARTITION BY rating ORDER BY length);
```

###  Max

MAX works in the same way as SUM

```python
%%sql

SELECT film_id,
        title,
        rating,
        length,
        MAX(length) OVER
            (PARTITION BY rating
             ORDER BY film_id)
            AS max
FROM film
WHERE title LIKE '%y';
```

(Below)

**"After we specify the field to order by, we add a definition for our window size: ROWS BETWEEN 3 PRECEDING AND 1 PRECEDING. This clause is essentially saying, "look back across the previous 3 films (not including the current order) and take the maximum value.""**

```python
%%sql

SELECT film_id,
        title,
        rating,
        length,
        MAX(length) OVER
            (PARTITION BY rating
             ORDER BY film_id
             ROWS BETWEEN 3 PRECEDING AND 1 PRECEDING)
            AS max
FROM film
WHERE title LIKE '%y';
```

# Common Table Expressions


"In PostgreSQL, the CTE(Common Table Expression) is used as a temporary result set that the user can reference within another SQL statement like SELECT, INSERT, UPDATE or DELETE. CTEs are temporary in the sense that they only exist during the execution of the query. CTEs are typically used to simplify complex joins and subqueries in PostgreSQL."

https://www.geeksforgeeks.org/postgresql-cte/


- Letâ€™s analyze the below syntax.

- The first step is to set the name of the CTE followed by an optional column list.

- Then we specify a query that returns the result set within the body of WITH clause. 

    - If not specified explicitly, then the select list of the CTE_query_definition will become the column list of the CTE.

- Finally use the CTE like a table or view in the statement which can be a SELECT, INSERT, UPDATE, or DELETE.

```python
%%sql

WITH cte_film AS (
    SELECT film_id,
            title,
            length,
            length > 120 AS over2hrs
    FROM film
)
SELECT film_id,
        title,
        length
FROM cte_film
WHERE over2hrs = TRUE
ORDER BY 1;
```

### CTE with case when

```python
%%sql

WITH cte_film AS (
    SELECT film_id, 
            title,
            (CASE 
                WHEN length < 90 THEN 'Short'
                WHEN length < 120 THEN 'Medium'
                ELSE 'Long'
            END) AS length    
    FROM film
)
SELECT film_id,
        title,
        length
FROM cte_film
WHERE length = 'Short'
ORDER BY 1;
```

### CTE with Window Functions

Finding the 2nd longest films in each age rating category using DENSE_RANK().

```python
%%sql

WITH cte_film AS  (
    SELECT film_id,
            title,
            rating,
            length,
            DENSE_RANK() OVER (
                PARTITION BY rating
                ORDER BY length DESC) 
            AS length_rank
    FROM film
)
SELECT *
FROM cte_film
WHERE length_rank = 2;
```
